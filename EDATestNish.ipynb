--- GRAPHICAL REPRESENTATIONS ---

```python
df['Cost'] = df['Sales'] - df['Profit']
```

```python
summary = pd.DataFrame({
    'Sales': [df['Sales'].sum()],
    'Cost': [df['Cost'].sum()],
    'Profit': [df['Profit'].sum()]
})

plt.figure(figsize=(6,4))
summary.T.plot(kind='bar')
plt.title("Total Sales vs Cost vs Profit")
plt.ylabel("Amount")
plt.xticks(rotation=0)
plt.show()
```

```python
sales_by_year = df.groupby(df['Order Date'].dt.year)['Sales'].sum()

plt.figure(figsize=(8,5))
sales_by_year.plot(kind='bar')
plt.title("Total Sales by Year")
plt.ylabel("Sales")
plt.show()
```

```python
profit_by_country = df.groupby('Country')['Profit'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
profit_by_country.head(10).plot(kind='bar')
plt.title("Top 10 Countries by Profit")
plt.ylabel("Profit")
plt.show()
```

```python
india_profit = df[df['Country'] == 'India']['Profit'].sum()
india_profit
```

```python
plt.figure(figsize=(5,4))
plt.bar(['India'], [india_profit])
plt.title("Total Profit in India")
plt.ylabel("Profit")
plt.show()
```
